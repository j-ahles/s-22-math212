#What did we learn?

We tried to model the position data, which initially had 8 different bounces.  On each bounce, the ball's height decreased.  We discussed three possibilities:

      1. A line is a bad fit - it won't capture the bounces.
      
      2. A quadratic could capture *one* of the bounces, but not all of them.

      3. A cosine function could capture the repeated bounces - but we'd need the amplitude to keep decreasing and we'd need to use absolute value to make the output always be positive.

image.png

We tried to fit the cosine function.  It didn't seem to work.  

Instructor's note 1:  This didn't work in class.  The first was that using (a2/t2) for the amplitude causes us to divide by 0.  So, to avoid this, use (a2 / [ t2 + b2 ] ).  

Instructor's note 2:  Also, the bounces don't take the same amount of time - so the function isn't really periodic.  That's why we didn't get a good fit! ]



image.png

A better fit was done by using just a quadratic function.  We erased the data in the table and just used the data from $t=0$ to the first bounce, which is at $t = 1.6$ seconds.  The quadratic function fit pretty well.
